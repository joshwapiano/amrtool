```python
import pandas as pd
import numpy as np
```

```python
# read in csv
amr_data = pd.read_csv("data/amr_dummy-csver.csv")
```

```python
# show head
amr_data.head()
```

```python
# create colour rules for each variable
def colour_rag_kno_a(val):
    colour = 'green'
    if type(val) is not str and val<0.7:
        colour = 'orange'
    if type(val) is not str and val<0.5:
        colour = 'red'
    return 'color: %s' % colour

def colour_rag_kno_b(val):
    colour = 'black'
    if type(val) is not str and val<0.7:
        colour = 'red'
    return 'color: %s' % colour
```

```python
# apply conditional formatting
a = amr_data.style.applymap(colour_rag_kno_a, subset = 'knowledge_a').applymap(colour_rag_kno_b, subset = 'knowledge_b')
a
```

```python
# subset output by Master row and village 2
amr_subset = amr_data.loc[amr_data['indicator'] == 'M'].loc[amr_data['village_group']==2]
```

```python
# apply conditional formatting
b = amr_subset.style.applymap(colour_rag_kno_a, subset = 'knowledge_a').applymap(colour_rag_kno_b, subset = 'knowledge_b')
b
```

```python
# Create table widget
import ipywidgets as widgets
from ipywidgets import interact
import IPython.display as IPdisp
```

```python
items = ['Please select:']+['All']+sorted(amr_data['village'].unique().tolist())
```

```python
# Diplay two tables, each with specific columns
```

```python
def view(Village=''):
    if Village=='Please select:': return
    if Village=='All':
        display_data = amr_data
    else:
        display_data = amr_data[amr_data['village']==Village]
    return IPdisp.display(\
                          display_data[['county','village','knowledge_a','antibiotics_a']].style.applymap(colour_rag_kno_a, subset = 'knowledge_a'),\
                          display_data[['county','village','knowledge_b','antibiotics_b']].style.applymap(colour_rag_kno_b, subset = 'knowledge_b'))
w = widgets.Select(options=items)
interact(view,Village=w); # note: by including ';' we remove the unwanted Out[] aspect.
```

```python
# Diplay two tables, each with specific columns, either simple or detailed
items = ['Please select:']+['All']+sorted(amr_data['village'].unique().tolist())
look_item = ['Detailed','Summarized']
```

```python
def look_tab(Village='',View='Detailed'):
    if Village=='Please select:': return
    if Village=='All':
        if View == 'Detailed':
            display_data = amr_data
        if View == 'Summarized':
            display_data = amr_data.loc[amr_data['indicator'] == 'M']
    
    
    else:
        if View == 'Detailed':
            display_data = amr_data[amr_data['village']==Village]
        if View == 'Summarized':
            display_data = amr_data[amr_data['village']==Village].loc[amr_data['indicator'] == 'M']
    return IPdisp.display(\
                          display_data[['county','village','knowledge_a','antibiotics_a']].style.applymap(colour_rag_kno_a, subset = 'knowledge_a'),\
                          display_data[['county','village','knowledge_b','antibiotics_b']].style.applymap(colour_rag_kno_b, subset = 'knowledge_b'))
w = widgets.Select(options=items)
v = widgets.RadioButtons(options=look_item, description='View type')
interact(look_tab,Village=w, View=v); # note: by including ';' we remove the unwanted Out[] aspect.
```

```python
# We can use interactive_output to display widgets side-by-side etc
```

```python
def look_tab(w='',v='Detailed'):
    if w=='Please select:': return
    if w=='All':
        if v == 'Detailed':
            display_data = amr_data
        if v == 'Summarized':
            display_data = amr_data.loc[amr_data['indicator'] == 'M']
    
    
    else:
        if v == 'Detailed':
            display_data = amr_data[amr_data['village']==w]
        if v == 'Summarized':
            display_data = amr_data[amr_data['village']==w].loc[amr_data['indicator'] == 'M']
    return IPdisp.display(\
                          display_data[['county','village','knowledge_a','antibiotics_a']].style.applymap(colour_rag_kno_a, subset = 'knowledge_a'),\
                          display_data[['county','village','knowledge_b','antibiotics_b']].style.applymap(colour_rag_kno_b, subset = 'knowledge_b'))

w = widgets.Select(options=items)
v = widgets.RadioButtons(options=look_item, description='View type')
ui = widgets.HBox([w,v])

out = widgets.interactive_output(look_tab, {'w': w, 'v': v})

display(ui, out)
```
